```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=3)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

UPDATE THE B VALUE WITH FORMULA -2*np.sum(y-m*X.ravel()-b)
here in the sum of formula y_cap = m*X+b so we are getting all the sum of all actual y-y_cap and finding the error mean square 

```python
m=93.55
b=100

loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
lr=0.1
step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1 = ((93.55 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 100')
plt.legend()
plt.show()
```

IF WE START FROM -100 AND WE CAN USE LOOP 

```python
b = -100
m = 93.55
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
